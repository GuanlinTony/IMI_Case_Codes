## Section 1. Data Limitations and AML implications

```python
datasets = {
    "abm": abm,
    "card": card,
    "cheque": cheque,
    "eft": eft,
    "emt": emt,
    "westernunion": westernunion,
    "wire": wire,
    "kyc_individual": kyc_individual,
    "kyc_smallbusiness": kyc_smallbusiness,
    "kyc_occupation_codes": kyc_occupation_codes,
    "kyc_industry_codes": kyc_industry_codes,
    "labels": labels
}

pd.DataFrame({
    name: {
        "rows": df.shape[0],
        "columns": df.shape[1]
    }
    for name, df in datasets.items()
}).T

## missing value percentages
missing_pct = {
    name: (df.isna().mean() * 100).round(2)
    for name, df in datasets.items()
}
missing_pct
```

```python
critical_cols = {
    "kyc_individual": ["income", "birth_date", "occupation_code", "country", "province", "city", "onboard_date"],
    "kyc_smallbusiness": ["sales", "employee_count", "industry_code", "country", "province", "city", "onboard_date"],
    "abm": ["amount_cad", "debit_credit", "transaction_datetime", "cash_indicator", "country", "province", "city"],
    "card": ["amount_cad", "debit_credit", "transaction_datetime", "merchant_category", "ecommerce_ind", "country", "province", "city"],
    "eft": ["amount_cad", "debit_credit", "transaction_datetime", "country", "province", "city"],
    "emt": ["amount_cad", "debit_credit", "transaction_datetime", "country", "province", "city"],
    "wire": ["amount_cad", "debit_credit", "transaction_datetime", "country", "province", "city"],
    "westernunion": ["amount_cad", "debit_credit", "transaction_datetime", "country", "province", "city"]
}

missing_critical = {}
for name, cols in critical_cols.items():
    df = datasets[name]
    cols_present = [c for c in cols if c in df.columns]
    missing_critical[name] = (df[cols_present].isna().mean() * 100).round(2)

missing_critical
```

```python
# unique customer counts
customer_counts = {}

for name, df in datasets.items():
    if "customer_id" in df.columns:
        customer_counts[name] = df["customer_id"].nunique()

pd.Series(customer_counts)
```

```python
all_customers_txn = pd.concat(
    [df[["customer_id"]] for df in [abm, card, cheque, eft, emt, westernunion, wire]],
    ignore_index=True
)["customer_id"].dropna().unique()

ind_customers = set(kyc_individual["customer_id"].dropna().unique())
bus_customers = set(kyc_smallbusiness["customer_id"].dropna().unique())
txn_customers = set(all_customers_txn)

coverage = pd.Series({
    "txn_customers_total": len(txn_customers),
    "txn_customers_in_individual_kyc": len(txn_customers & ind_customers),
    "txn_customers_in_business_kyc": len(txn_customers & bus_customers),
    "txn_customers_in_both_kyc": len(txn_customers & ind_customers & bus_customers),
    "txn_customers_missing_kyc": len(txn_customers - (ind_customers | bus_customers))
})
coverage
```

```python
# Check duplicate customer IDs in kyc_individual and kyc_smallbusiness

duplicate_txn = {}

for name, df in datasets.items():
    if "transaction_id" in df.columns:
        duplicate_txn[name] = df.duplicated("transaction_id").sum()

pd.Series(duplicate_txn)
```

```python
txn_tables = {
    "abm": abm,
    "card": card,
    "cheque": cheque,
    "eft": eft,
    "emt": emt,
    "westernunion": westernunion,
    "wire": wire
}

dq = {}

for name, df in txn_tables.items():
    dq[name] = {
        "rows": len(df),
        "missing_customer_id": int(df["customer_id"].isna().sum()),
        "missing_transaction_id": int(df["transaction_id"].isna().sum()),
        "missing_datetime": int(df["transaction_datetime"].isna().sum()),
        "nonpositive_amount": int((df["amount_cad"] <= 0).sum())
    }

pd.DataFrame(dq).T
```

```python
# transactions amount summary statistics
txn_tables = {
    "abm": abm,
    "card": card,
    "cheque": cheque,
    "eft": eft,
    "emt": emt,
    "westernunion": westernunion,
    "wire": wire
}

amount_stats = {
    name: df["amount_cad"].describe()
    for name, df in txn_tables.items()
}

pd.DataFrame(amount_stats).T
```

```python
quantiles = [0.5, 0.9, 0.95, 0.99, 0.999]
q_table = {}

for name, df in txn_tables.items():
    q_table[name] = df["amount_cad"].quantile(quantiles)

pd.DataFrame(q_table).T
```

```python
# debit_credit amount summary statistics (Debit: D, Credit: C)
debit_credit_amounts = {
    name: df.groupby("debit_credit")["amount_cad"].describe()
    for name, df in txn_tables.items()
}

debit_credit_amounts
```

```python
for name, df in txn_tables.items():
    plt.figure()
    sns.histplot(df["amount_cad"], bins=100)
    plt.title(f"{name.upper()} Transaction Amount Distribution")
    plt.xlabel("Amount (CAD)")
    plt.ylabel("Count")
    plt.show()
# We use a log scale for transaction amounts because they are highly right-skewed with a few very large values. 
# we use log(x+1) because this handles zero amounts without issues and preserves the order of values, welcome to adjust it when reasonable.
for name, df in txn_tables.items():
    plt.figure()
    sns.histplot(np.log1p(df["amount_cad"]), bins=100)
    plt.title(f"{name.upper()} Log(Transaction Amount + 1)")
    plt.xlabel("log(Amount + 1)")
    plt.ylabel("Count")
    plt.show()

```

```python
all_txns = []

for name, df in txn_tables.items():
    temp = df[["amount_cad"]].copy()
    temp["channel"] = name
    all_txns.append(temp)

amount_channel_df = pd.concat(all_txns, ignore_index=True)
plt.figure(figsize=(12,6))
sns.boxplot(data=amount_channel_df, x="channel", y="amount_cad")
plt.yscale("log")
plt.title("Transaction Amount Distribution by Channel (Log Scale)")
plt.xlabel("Channel")
plt.ylabel("Amount (CAD)")
plt.show()
```

```python
# transaction amount concentration in top 1% transactions
concentration = {}

for name, df in txn_tables.items():
    threshold = df["amount_cad"].quantile(0.99)
    concentration[name] = (
        df.loc[df["amount_cad"] >= threshold, "amount_cad"].sum()
        / df["amount_cad"].sum()
        # This percentage measures how much of the total transaction value is contributed by the largest 1% of transactions.
    )

pd.Series(concentration)
```

## Section 3. KYC-related transaction behavior mismatches

```python
# KYC-linked transaction behavior refers to analyzing transaction patterns in the context of a customers KYC profile.
# behavior that deviates from what the KYC profile would suggest might be indicative of suspicious activity in AML.
txn_tables = {
    "abm": abm,
    "card": card,
    "cheque": cheque,
    "eft": eft,
    "emt": emt,
    "westernunion": westernunion,
    "wire": wire
}

txns = []

for channel, df in txn_tables.items():
    temp = df[["customer_id", "amount_cad", "debit_credit"]].copy()
    temp["channel"] = channel
    txns.append(temp)

transactions_all = pd.concat(txns, ignore_index=True)

```

```python
# Aggregare transaction data at customer level
customer_txn_features = transactions_all.groupby("customer_id").agg(
    txn_count=("amount_cad", "count"),
    total_amount=("amount_cad", "sum"),
    avg_amount=("amount_cad", "mean"),
    max_amount=("amount_cad", "max"),
    channel_count=("channel", "nunique")
).reset_index()
```

```python
# Channel usage features
channel_usage = (
    transactions_all
    .groupby(["customer_id", "channel"])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)
```

```python
# Merge with individual KYC data
individual_features = (
    customer_txn_features
    .merge(channel_usage, on="customer_id", how="left")
    .merge(kyc_individual, on="customer_id", how="inner")
)

# Merge with small business KYC data
business_features = (
    customer_txn_features
    .merge(channel_usage, on="customer_id", how="left")
    .merge(kyc_smallbusiness, on="customer_id", how="inner")
)
```

```python
## KYC-customer type

kyc_type = pd.DataFrame({"customer_id": list(ind_customers | bus_customers)})
kyc_type["customer_type"] = np.where(
    kyc_type["customer_id"].isin(ind_customers) & kyc_type["customer_id"].isin(bus_customers), "both",
    np.where(kyc_type["customer_id"].isin(ind_customers), "individual",
             np.where(kyc_type["customer_id"].isin(bus_customers), "business", "unknown"))
)

txn_customer_base = customer_txn_features.merge(kyc_type, on="customer_id", how="left")
txn_customer_base["customer_type"] = txn_customer_base["customer_type"].fillna("unknown")

txn_customer_base["customer_type"].value_counts()
```

```python
# KYC-linked transaction behavior refers to analyzing transaction patterns in the context of a customers KYC profile.
# visualizations

# 1- Income vs Transaction Behavior
# Individauls
individual_features[["income", "total_amount", "txn_count", "channel_count"]].describe()

plt.figure(figsize=(6,4))
plt.scatter(individual_features["income"], individual_features["total_amount"], alpha=0.3)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Declared Income (log)")
plt.ylabel("Total Transaction Amount (log)")
plt.title("Individual Income vs Transaction Volume")
plt.show()

# 2 - Small Business
business_features[["sales", "total_amount", "txn_count", "employee_count"]].describe()

plt.figure(figsize=(6,4))
plt.scatter(business_features["sales"], business_features["total_amount"], alpha=0.3)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Declared Sales (log)")
plt.ylabel("Total Transaction Amount (log)")
plt.title("Business Sales vs Transaction Volume")
plt.show()

```

```python
# Simple mismatch indicators (feature intuition)

individual_features["amount_to_income_ratio"] = (
    individual_features["total_amount"] / individual_features["income"]
)

business_features["amount_to_sales_ratio"] = (
    business_features["total_amount"] / business_features["sales"]
)

individual_features["amount_to_income_ratio"].describe()
business_features["amount_to_sales_ratio"].describe()
```

```python
# Peer-based mismatch
individual_features["occupation_code"] = individual_features["occupation_code"].astype("string").str.strip()
kyc_occupation_codes["occupation_code"] = kyc_occupation_codes["occupation_code"].astype("string").str.strip()
ind_peer = individual_features.merge(kyc_occupation_codes, on="occupation_code", how="left")
ind_peer["peer_total_amount_median"] = ind_peer.groupby("occupation_code")["total_amount"].transform("median")
ind_peer["total_vs_peer_median"] = ind_peer["total_amount"] / ind_peer["peer_total_amount_median"]

business_features["industry_code"] = business_features["industry_code"].astype("string").str.strip()
kyc_industry_codes["industry_code"] = kyc_industry_codes["industry_code"].astype("string").str.strip()
bus_peer = business_features.merge(kyc_industry_codes, on="industry_code", how="left")
bus_peer["peer_total_amount_median"] = bus_peer.groupby("industry_code")["total_amount"].transform("median")
bus_peer["total_vs_peer_median"] = bus_peer["total_amount"] / bus_peer["peer_total_amount_median"]

ind_peer[["occupation_code", "occupation_title", "total_vs_peer_median"]].sort_values("total_vs_peer_median", ascending=False).head(20)
bus_peer[["industry_code", "industry", "total_vs_peer_median"]].sort_values("total_vs_peer_median", ascending=False).head(20)
```

## Section 4: Behavioral clusters (labeled-agnostic)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

customer_features = customer_txn_features.merge(channel_usage, on="customer_id", how="left")
customer_features = customer_features.merge(kyc_type[["customer_id", "customer_type"]], on="customer_id", how="left")
customer_features["customer_type"] = customer_features["customer_type"].fillna("unknown")

customer_features = customer_features.merge(labels[["customer_id", "label"]], on="customer_id", how="left")
customer_features["is_labeled"] = customer_features["label"].notna()
```

```python
# Merge the datasets
dc = (
    transactions_all.groupby(["customer_id", "debit_credit"])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

customer_features = customer_features.merge(dc, on="customer_id", how="left")

if "D" in customer_features.columns and "C" in customer_features.columns:
    customer_features["debit_share"] = customer_features["D"] / (customer_features["D"] + customer_features["C"]).replace(0, np.nan)
else:
    customer_features["debit_share"] = np.nan
```

```python
exclude_cols = {"customer_id", "customer_type", "label", "is_labeled"}
num_cols = [c for c in customer_features.columns if c not in exclude_cols and pd.api.types.is_numeric_dtype(customer_features[c])]

X = customer_features[num_cols].copy()

for col in ["total_amount", "avg_amount", "max_amount"]:
    if col in X.columns:
        X[col] = np.log1p(X[col])

X = X.fillna(0)
```

```python
# Correlation analysis of numericals with heatmap
corr = X.corr()

plt.figure(figsize=(12, 10))
plt.imshow(corr, aspect="auto")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap (Customer-level numeric features)")
plt.tight_layout()
plt.show()
```

```python
# PCA for dimensionality reduction
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

pca.explained_variance_ratio_
```

```python
# Plot PCA results (no clustering)
plt.figure(figsize=(7, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.25)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA of Customer Behavioral Features (Unclustered)")
plt.tight_layout()
plt.show()
```

```python
# KMeans - Inertia
ks = range(2, 11)
inertias = []

for k in ks:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_scaled)
    inertias.append(km.inertia_)

plt.figure(figsize=(7, 5))
plt.plot(list(ks), inertias, marker="o")
plt.xlabel("k")
plt.ylabel("Inertia")
plt.title("Elbow Plot for KMeans")
plt.tight_layout()
plt.show()
```

```python
# KMeans - Silhouette Score
sil_scores = []

for k in ks:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels_k = km.fit_predict(X_scaled)
    sil_scores.append(silhouette_score(X_scaled, labels_k))

plt.figure(figsize=(7, 5))
plt.plot(list(ks), sil_scores, marker="o")
plt.xlabel("k")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for KMeans")
plt.tight_layout()
plt.show()
```

```python
# Fit KMeans with selected k  and visualize PCA sactter colored by clusters

k = 3
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
customer_features["cluster"] = kmeans.fit_predict(X_scaled)
customer_features["cluster"].value_counts()

plt.figure(figsize=(7, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=customer_features["cluster"], alpha=0.35)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Behavioral Clusters (KMeans) shown in PCA space")
plt.tight_layout()
plt.show()
```

```python
# Overlay labels on clusters
# to see if patterns extend beyond labeled customers

mask_labeled = customer_features["is_labeled"].values

plt.figure(figsize=(7, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=customer_features["cluster"], alpha=0.20)
plt.scatter(X_pca[mask_labeled, 0], X_pca[mask_labeled, 1], alpha=0.9)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clusters with Labeled Customers Overlaid")
plt.tight_layout()
plt.show()

```

```python
# Cluster profiling table for AML libarary
profile_cols = ["txn_count", "total_amount", "avg_amount", "max_amount", "channel_count", "debit_share"]
profile_cols = [c for c in profile_cols if c in customer_features.columns]

cluster_profile = customer_features.groupby("cluster")[profile_cols].median()
cluster_sizes = customer_features.groupby("cluster").size().rename("n")
labeled_rate = customer_features.groupby("cluster")["is_labeled"].mean().rename("pct_labeled")

pd.concat([cluster_sizes, labeled_rate, cluster_profile], axis=1).sort_values("n", ascending=False)
```

```python
# Check channel mix by cluster in typologies

channel_cols = [c for c in ["abm","card","cheque","eft","emt","westernunion","wire"] if c in customer_features.columns]

cluster_channel = customer_features.groupby("cluster")[channel_cols].mean()

plt.figure(figsize=(10, 5))
plt.imshow(cluster_channel.values, aspect="auto")
plt.colorbar()
plt.xticks(range(len(channel_cols)), channel_cols, rotation=45)
plt.yticks(range(cluster_channel.shape[0]), cluster_channel.index)
plt.title("Average Channel Usage by Cluster")
plt.tight_layout()
plt.show()

cluster_channel
```
